```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("evaluation.csv")
```

```python
df
```

```python



# Plotting
plt.figure(figsize=(12, 6))
for model in ["llama_3.2_1b_r8_BLEU", "llama_3.2_1b_r16_BLEU", "llama_3.2_1b_r32_BLEU", "phi-3.5-mini_BLEU","rag_BLEU"]:
    plt.plot(df["Unnamed: 0.2"], df[model], marker='o', label=model)

plt.title("BLEU Scores Across Testing Items for Models", fontsize=14)
plt.xlabel("Item", fontsize=12)
plt.ylabel("BLEU Score", fontsize=12)
plt.legend(title="Models")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the graph
models = [
    "LLaMA-3.2-1b (R=8)",
    "LLaMA-3.2-1b (R=16)",
    "LLaMA-3.2-1b (R=32)",
    "Phi 3.5 Mini",
    "RAG"
]
parameters = [12156928, 24313856, 48627712, 29884416, 0]

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(models, parameters, marker='o', linestyle='-', color='b', label='Trained Parameters')

# Adding titles and labels
plt.title("Parameters Trained for Different Models", fontsize=16)
plt.xlabel("Model", fontsize=14)
plt.ylabel("Trained Parameters", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)

# Annotating data points for clarity
for i, value in enumerate(parameters):
    plt.text(i, value + 1e6, f"{value:,}", ha='center', fontsize=10, color='darkblue')

# Show legend and graph
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
